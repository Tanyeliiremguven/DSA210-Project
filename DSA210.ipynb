```python
# Ensure the required library is installed
try:
    from pptx import Presentation
    from pptx.util import Inches, Pt
except ModuleNotFoundError:
    import os
    os.system("pip install python-pptx")
    from pptx import Presentation
    from pptx.util import Inches, Pt

# Create a presentation object
prs = Presentation()

# Slide 1: Title Slide
title_slide = prs.slides.add_slide(prs.slide_layouts[0])
title = title_slide.shapes.title
title.text = "Steps Analysis During Exam Periods"
subtitle = title_slide.placeholders[1]
subtitle.text = "Understanding the Impact of Exam Periods on Physical Activity\nYour Name\nDSA210 - January 2025"

# Slide 2: Introduction
slide = prs.slides.add_slide(prs.slide_layouts[1])
title = slide.shapes.title
title.text = "Introduction"
content = slide.placeholders[1]
content.text = (
    "This project analyzes daily step counts to understand how physical activity changes during exam periods. "
    "It compares step counts before, during, and after exams to regular days, aiming to identify patterns and correlations "
    "that reflect the impact of stress and time constraints."
)

# Slide 3: Data Sources
slide = prs.slides.add_slide(prs.slide_layouts[1])
title = slide.shapes.title
title.text = "Data Sources"
content = slide.placeholders[1]
content.text = (
    "1. Steps Data:\n"
    "- Extracted from the Health app on an iPhone.\n"
    "- Cleaned, processed, and saved as a CSV file.\n\n"
    "2. Exam Dates Data:\n"
    "- Manually compiled from university schedules.\n"
    "- Recorded in a CSV file with dates formatted to match step data."
)

# Slide 4: Methodology
slide = prs.slides.add_slide(prs.slide_layouts[1])
title = slide.shapes.title
title.text = "Methodology"
content = slide.placeholders[1]
content.text = (
    "- Collected and cleaned steps and exam dates data.\n"
    "- Filtered steps data to include records from 2023 onwards.\n"
    "- Categorized days as Before Exam, On Exam, After Exam, or Regular Day.\n"
    "- Performed statistical analysis and visualized results."
)

# Slide 5: Statistical Analysis
slide = prs.slides.add_slide(prs.slide_layouts[1])
title = slide.shapes.title
title.text = "Statistical Analysis"
content = slide.placeholders[1]
content.text = (
    "- Mean Step Counts:\n"
    "  * Exam Periods: 5789.70 steps/day\n"
    "  * Regular Days: 6937.02 steps/day\n\n"
    "- Regression Analysis:\n"
    "  * Exam Periods: y1 = -19.24x + 17463.90\n"
    "  * Regular Days: y2 = -7.49x + 9433.53"
)

# Slide 6: Visualizations
slide = prs.slides.add_slide(prs.slide_layouts[1])
title = slide.shapes.title
title.text = "Visualizations"
content = slide.placeholders[1]
content.text = (
    "- Box Plot: Compares step counts during exam periods vs. regular days.\n"
    "- Time-Series Plot: Displays daily step counts with exam dates marked."
)

# Slide 7: Findings
slide = prs.slides.add_slide(prs.slide_layouts[1])
title = slide.shapes.title
title.text = "Findings"
content = slide.placeholders[1]
content.text = (
    "- Step counts significantly decrease during exam periods.\n"
    "  * Average during exam periods: 5789.70 steps/day.\n"
    "  * Average on regular days: 6937.02 steps/day.\n\n"
    "- Variability in step counts is narrower during exams.\n"
    "- Stress and time constraints are likely contributors to reduced activity."
)

# Slide 8: Limitations and Future Work
slide = prs.slides.add_slide(prs.slide_layouts[1])
title = slide.shapes.title
title.text = "Limitations and Future Work"
content = slide.placeholders[1]
content.text = (
    "Limitations:\n"
    "- Focused only on step counts; excluded other health metrics.\n"
    "- Dataset limited to 2024 exam periods.\n\n"
    "Future Work:\n"
    "- Include additional metrics like sleep and stress levels.\n"
    "- Analyze data over multiple years for broader insights.\n"
    "- Test interventions to maintain activity during exams."
)

# Slide 9: Conclusion
slide = prs.slides.add_slide(prs.slide_layouts[1])
title = slide.shapes.title
title.text = "Conclusion"
content = slide.placeholders[1]
content.text = (
    "- Exam periods significantly impact physical activity.\n"
    "- Maintaining physical activity during exams is vital for overall well-being.\n"
    "- Recommendations:\n"
    "  * Take short walks or schedule light exercise during exams.\n"
    "  * Develop routines to balance study and physical activity."
)

# Save the presentation
prs.save("Steps_Analysis_Presentation.pptx")

print("Presentation created and saved as 'Steps_Analysis_Presentation.pptx'.")
```
